```python
import matplotlib.pyplot as plt
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
w = 1.0
```

```python
#forward 
def forward(x):
    return x * w
```

```python
# Loss function
def loss(x, y):
    y_pred = forward(x)
    return pow((y_pred - y),2)
```

```python
# compute gradient
def gradient(x, y):
    return 2 * x * (x * w - y)
```

```python
#backward
def backward(x,y):
    return w - 0.01 *gradient(x,y)
```

```python
# Training loop
arrLoss=[]
print(forward(4))
for epoch in range(100):
    for x_val, y_val in zip(x_data, y_data):
        w =  backward(x_val, y_val)
        print("\tgrad: ", x_val, y_val, w)
        l = loss(x_val, y_val)
    print("progress:", epoch, "w=", w, "loss=", l)
    arrLoss.append(l)
print(forward(4))
plt.plot(arrLoss)
plt.xlabel('W')
plt.ylabel('loss')
plt.show
```
